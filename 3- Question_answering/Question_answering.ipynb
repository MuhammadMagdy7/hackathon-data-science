```python
from haystack.document_stores import InMemoryDocumentStore

document_store = InMemoryDocumentStore()
```

```python
import pandas as pd

df = pd.read_csv('quesation.csv')
df
```

```python
df.drop_duplicates(subset='0' , keep='first', ignore_index=True, inplace= True)
df
```

```python
question = []
```

```python
df.dropna(axis=0, inplace= True)
```

```python
for i in df['0']:
  question.append(i)
```

```python
with open('text.txt') as f:
    text = f.read()
```

```python
object1 = [{"content": text}]
document_store.write_documents(object1)
```

```python
# An in-memory TfidfRetriever based on Pandas dataframes
from haystack.nodes import TfidfRetriever

retriever = TfidfRetriever(document_store=document_store)
```

```python
from haystack.nodes import FARMReader


# Load a  local model or any of the QA models on
# Hugging Face's model hub (https://huggingface.co/models)
reader = FARMReader(model_name_or_path="deepset/deberta-v3-large-squad2")
```

```python
d = {}
```

```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline

model_name = "deepset/deberta-v3-large-squad2"
# a) Get predictions
for i in range(30):
    nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
    QA_input = {
        'question': question[i],
        'context': text }
    res = nlp(QA_input)
    d[question[i]] = res['answer']

# b) Load model & tokenizer
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```
