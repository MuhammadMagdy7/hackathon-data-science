#### Create column for length NUN in each row

```python
index = df.index
addlist = []
for i in index:
    addlist.append(df.loc[[i]].isna().sum().sum())
```

```python
df['add'] = addlist
```

```python
df['add'].describe()
```

#### Drop each row has more 25 NUN

```python
drop = df[df['add'] > 25].index
```

```python
df.drop(drop, axis= 0, inplace=True)
```

```python
df.isna().sum().tail(55)
```

```python
years = df.drop(columns=['Area Code', 'Months Code', 'Element Code','Unit', 'add'], axis = 1)
```

```python
years
```

--------------------------

```python
removed = years[['Area', 'Months', 'Element']]
```

```python
years.drop(['Area', 'Months', 'Element'], axis= 1, inplace = True)
```

#### Fill nan in row by mean of row 

```python
index = years.index

for i in index:
    years.loc[i].fillna(float(format(years.loc[i].mean(),".3f")), inplace= True)
```

```python
years = pd.concat([removed, years], axis= 1)
```

```python
years.info()
```

```python
Temperature = years[years['Element'] == 'Temperature change']
```

```python
Standard_Deviation = years[years['Element'] == 'Standard Deviation']
```

```python
Standard_Deviation.drop('Element',axis= 1, inplace=True)
```

```python
Standard_Deviation
```

```python
months = Standard_Deviation.drop('Area', axis= 1)
```

```python
months = months.groupby(by='Months').agg('mean').reset_index().T
```

```python
tem_df = pd.read_csv('Temperature_change_Data.csv')
```

```python
tem_df.drop('Country Code', axis= 1, inplace=True)
```

```python
last_ten_years = tem_df[tem_df['year']>= 2009]
```

### What are the ten countries that suffer from temperature change at the very least in the last ten years?

```python
last_ten_years.groupby('Country Name').mean()['tem_change'].sort_values().head(10)
```

### What are the ten most countries that suffer from temperature change mostly in the last ten years?

```python
last_ten_years.groupby('Country Name').mean()['tem_change'].sort_values(ascending=False).head(10)
```
